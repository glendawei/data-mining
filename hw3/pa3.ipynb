```python
texts = []

for i in range(1, 1096):
    with open(os.path.join(document_folder, f"{i}.txt"), 'r', encoding='utf-8') as file:
        content = file.read()
        texts.append(content)
```

```python

from keras_bert import extract_embeddings


model_path = '/content/drive/My Drive/uncased_L-2_H-128_A-2'
dict_path = '/content/drive/My Drive/uncased_L-2_H-128_A-2/vocab.txt'

embeddings = extract_embeddings(model_path, texts)
print(len(embeddings))
print(embeddings)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np



# Rest of your code remains the same
x_train, y_train = [], []

docs_in_train = []
for i in classes:
    cls = i[0]
    for doc_id in i[1:]:
        x_train.append(embeddings[doc_id - 1][0])
        y_train.append(cls)
        docs_in_train.append(doc_id)
x_test, y_test = [], []
all_docs = set(range(1, 1096))

#print(docs_in_train)

docs_not_in_train = all_docs - set(docs_in_train)

for doc_id in docs_not_in_train:
    x_test.append(embeddings[doc_id - 1][0])
    #y_test.append(labels[doc_id - 1])
import numpy as np

# Your previous code for creating x_train, y_train, x_test, and y_test...




```

```python
from sklearn.svm import SVC
import csv
SVM_model = SVC(kernel='linear', C=1.0, probability=True)

SVM_model.fit(x_train,y_train)

predicted_results = []
expected_results = []

expected_results.extend(y_test)
predicted_results.extend(SVM_model.predict(x_test))
print(predicted_results)

with open("/content/drive/My Drive/op.csv", "w") as csv_file:
    writer = csv.writer(csv_file)
    writer.writerow(["Id", "Value"])

    # Convert set to list
    docs_not_in_train_list = list(docs_not_in_train)

    for i, doc_id in enumerate(docs_not_in_train_list):
        writer.writerow([doc_id, predicted_results[i]])
```
